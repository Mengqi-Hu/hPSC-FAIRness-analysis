```python
import matplotlib.pyplot as plt

# Total values for width calculation
registered_total = only_registered + both
published_total = only_published + both

# Normalize for visual scaling
unit = 5  # 1 unit = 5 items (you can tune this)
reg_width = registered_total / unit
pub_width = published_total / unit
height = 2  # fixed height

# Set up figure
fig, ax = plt.subplots(figsize=(8, 3))
ax.axis('off')

# Draw rectangles
# Rectangle(left, bottom, width, height)
reg_rect = Rectangle((0, 1), reg_width, height, facecolor='#1D4A73', alpha=0.6, edgecolor='black', label='Registered')
pub_rect = Rectangle((reg_width - (both / unit), 1), pub_width, height, facecolor='cadetblue', alpha=0.6, edgecolor='black', label='Published')

ax.add_patch(reg_rect)
ax.add_patch(pub_rect)

# Text labels
ax.text(reg_width / 3, 2.6, f'Registered only: {only_registered}', ha='center', fontsize=10)
ax.text(reg_width + (pub_width / 3) - (both / unit), 2.6, f'Published only: {only_published}', ha='center', fontsize=10)
ax.text(reg_width - (both / (2*unit)), 1.5, f'Both: {both}', ha='center', fontsize=10, fontweight='bold')

# Legend and title
ax.legend(loc='lower center', fontsize=10, frameon=False)
plt.title('Rectangular Venn: Registered vs Published', fontsize=12, fontweight='bold')
plt.xlim(0, reg_width + pub_width)
plt.ylim(0, 4)

plt.tight_layout()
plt.show()
```
